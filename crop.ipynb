```python
df_area_harvested
# df_yeild

years = df_area_harvested['Year'].unique()

totalCropAreaHarvested = []

for eachYear in years:
    df_year_area_harvested = df_area_harvested.loc[df_area_harvested['Year'] == eachYear]
    totalCropAreaHarvested.append(df_year_area_harvested['Value'].sum())

# converting hectare area to million hectare area
totalCropAreaHarvested = np.array(totalCropAreaHarvested)
totalCropAreaHarvested = totalCropAreaHarvested / 1000000

df_crop_area_overall = pd.DataFrame({
    'year': years, 
    'total_area_harvested': totalCropAreaHarvested
})

plt.scatter(df_crop_area_overall['year'], df_crop_area_overall['total_area_harvested'])
plt.show()
```
